# **PCA**

```python
pca2_x = df.copy()
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler


pca2_y = pca2_x.Default
pca2_x.drop(columns='Default')
pca2_x = pca2_x
pca2_x_norm = MinMaxScaler().fit_transform(pca2_x)
pca2_x_norm
```

```python
comps = 14
comp_cols = ["component_"+str(i+1) for i in range(comps)]
comp_cols
```

```python
from sklearn.decomposition import PCA
```

```python
pca2 = PCA(n_components=comps)
res2 = pca2.fit_transform(pca2_x_norm)
res2
```

```python
pca2_dataset = pd.DataFrame(data =res2 , columns = comp_cols )
pca2_dataset["Default"] = df.Default
pca2_dataset.shape
```

```python
import xgboost as xgb
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
x_p_1_train,x_p_1_test,Y_p_1_train,Y_p_1_test=train_test_split(pca2_dataset.drop(labels=['Default'], axis=1),
     pca2_dataset['Default'],
     test_size=0.3,
     random_state=0)

xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42,learning_rate=0.01, n_estimators=200, max_depth=3)
xgb_model.fit(x_p_1_train, Y_p_1_train)
preds_p_1=xgb_model.predict(x_p_1_test)

fs=f1_score(Y_p_1_test,preds_p_1,average='weighted')
print("score",fs)
```
